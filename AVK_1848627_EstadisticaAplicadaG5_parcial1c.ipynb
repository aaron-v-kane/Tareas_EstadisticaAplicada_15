```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrorrey = pd.read_csv('metrorrey.csv')
```

```python
metrorrey.plot()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

metrorrey['TESadd12'] = ExponentialSmoothing(metrorrey['Pasajeros'],trend='add',seasonal='add',seasonal_periods=12).fit().fittedvalues
metrorrey['TESmul12'] = ExponentialSmoothing(metrorrey['Pasajeros'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
```

```python
metrorrey[['Pasajeros','TESadd12','TESmul12']].plot(figsize=(12,10))
```
