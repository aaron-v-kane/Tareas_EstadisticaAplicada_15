***Los modelos AR(2) y la parte Autorregresiva del modelo ARMA(2,2) se hicieron con phi's negativas, dado que con phi's positivas el proceso no se simulaba de manera adecuada***

```python
#empezamos importando las librerias
import numpy as np
import pandas as pd 
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_process import ArmaProcess

%matplotlib inline
```

```python
#fijamos tamaño de graficas
plt.rcParams['figure.figsize'] = [10, 7.5]
```

**Primer Modelo: MA(2)**

$Y_t =  \varepsilon_t +0.7 \varepsilon_{t-1} + 0.2 \varepsilon_{t-2} $ 

```python
#generamos el modelo
np.random.seed(29)
ma2 = np.array([1, 0.7, 0.2])
ar1 = np.array([1])
proceso_MA2 = ArmaProcess(ar1, ma2).generate_sample(1000)
```

```python
#graficamos el modelo
plt.plot(proceso_MA2)
plt.title(r'Proceso de medias móviles de 2do orden $Y_t= \varepsilon_t + 0.7\varepsilon_{t-1}+0.2\varepsilon_{t-2}$ ')
plt.show()
```

```python
#Realizamos la prueba de Raíces Unitarias (ADF)
result_adf = adfuller(proceso_MA2)
print('p-value: %f' % result_adf[1])
```

 **$H_0:$ serie no estacionaria**

 **$H_1:$ serie estacionaria**


Dado que p-value < α, se rechaza H0, por lo que la serie es estacionaria

```python
#Graficamos su fuinción de autocorrelación (ACF)
plot_acf(proceso_MA2, zero=False, lags=30)
plt.title(r'autocorrelación MA(2) $Y_t = \varepsilon_t + 0.7 \varepsilon_{t-1}+ 0.2 \varepsilon_{t-2}$ ')
plt.show()
```

```python
#La de autocorrelación parcial (PACF)
plot_pacf(proceso_MA2, zero=False, lags=30)
plt.title(r'autocorrelación parcial MA(2) $Y_t = \varepsilon_t + 0.7 \varepsilon_{t-1}+ 0.2 \varepsilon_{t-2}$')
plt.show() 
```

**Segundo Modelo: AR(2)**

$Y_t =  -0.8 Y_{t-1} - 0.6 Y_{t-2} $

```python
#generamos el modelo
np.random.seed(39)
ma1 = np.array([1])
ar2 = np.array([1, 0.8, 0.6])
proceso_AR2 = ArmaProcess(ar2, ma1).generate_sample(1000)
```

```python
#graficamos el modelo
plt.plot(proceso_AR2)
plt.title(r'Proceso autoregresivo de 2do orden $Y_t =  -0.8 Y_{t-1} - 0.6 Y_{t-2} +\varepsilon_t$ ')
plt.show()
```

```python
#Realizamos la prueba de Raíces Unitarias (ADF)
result_adf = adfuller(proceso_AR2)
print('p-value: %f' % result_adf[1])
```

**$H_0:$ serie no estacionaria**

 **$H_1:$ serie estacionaria**


Dado que p-value < α, se rechaza H0, por lo que la serie es estacionaria

```python
#Graficamos su fuinción de autocorrelación (ACF)
plot_acf(proceso_AR2, zero=False, lags=30)
plt.title(r'autocorrelación AR(2) $Y_t =  -0.8 Y_{t-1} - 0.6 Y_{t-2} +\varepsilon_t$ ')
plt.show()
```

```python
#La de autocorrelación parcial (PACF)
plot_pacf(proceso_AR2, zero=False, lags=30)
plt.title(r'autocorrelación parcial AR(2) $Y_t =  -0.8 Y_{t-1} - 0.6 Y_{t-2} +\varepsilon_t$')
plt.show() 
```

**Tercer Modelo: ARMA(2,2)**

```python
#Realizamos la prueba de Raíces Unitarias (ADF)
result_adf = adfuller(proceso_ARMA2)
print('p-value: %f' % result_adf[1])
```

**$H_0:$ serie no estacionaria**

 **$H_1:$ serie estacionaria**


Dado que p-value < α, se rechaza H0, por lo que la serie es estacionaria

```python
#Graficamos su fuinción de autocorrelación (ACF)
plot_acf(proceso_ARMA2, zero=False, lags=30)
plt.title(r'autocorrelación ARMA(2,2) $Y_t =  0.7 \varepsilon_{t-1} + 0.2 \varepsilon_{t-2} - 0.8 Y_{t-1} - 0.6 Y_{t-2} + \varepsilon_t$ ')
plt.show()
```
